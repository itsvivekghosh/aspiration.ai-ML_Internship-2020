__CAPM (Capital Asset Pricing Model): CAPM Analysis and Beta Calculation using regression -__

- CAPM(Capital Asset Pricing Model) attempts to price securities by examining the relationship that exists between expected returns and risk. ***The Capital Asset Pricing Model (CAPM) describes the relationship between systematic risk and expected return for assets, particularly stocks. CAPM is widely used throughout finance for pricing risky securities and generating expected returns for assets given the risk of those assets and cost of capital. ***

__Read more about CAPM. <a href="https://www.investopedia.com/terms/c/capm.asp">(Investopedia CAPM reference)__</a>

- The Beta of an asset is a measure of the sensitivity of its returns relative to a market benchmark (usually a market index). How sensitive/insensitive is the returns of an asset to the overall market returns (usually a market index like S&P 500 index). What happens when the market jumps, does the returns of the asset jump accordingly or jump somehow? ***Beta is a measure of a stock's volatility in relation to the overall market. By definition, the market, such as the S&P 500 Index, has a beta of 1.0, and individual stocks are ranked according to how much they deviate from the market. ***

__Read more about Beta <a href="https://www.investopedia.com/investing/beta-know-risk/">(Investopedia Beta reference)</a>__

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
plt.style.use("ggplot")
```

```python
df = pd.read_csv("Datasets/GOLD.csv", date_parser=['Date'])
df.Date = pd.to_datetime(df.Date)
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(df['Pred'], df["new"], s=250, marker='.', color = 'green')
plt.xlabel("Pred", fontsize=14)
plt.ylabel("New", fontsize=14)
plt.title("Plot for Pred v/s New", fontsize=17)
plt.show()
```

```python
X = df['Pred'].values
Y = df['new'].values
```

```python
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, Y, test_size=0.1, random_state=100)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
# Model Evaluation Metrics

rmse = mean_squared_error(y_test, y_pred)

r2 = r2_score(y_test, y_pred)

print("Mean Squared Error of the Model is:", rmse)
print('R^2 Score of the Model is:', r2)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, marker="o", label = 'New vs Pred', s=50, color = 'blue')
plt.plot(X_test, y_pred, 'r', label = 'Regression Line')
plt.xlabel('New Values', fontsize=15)
plt.ylabel('Pred Values', fontsize=15)
plt.title("Plot for New Values with Predicted Values", fontsize=18)
plt.legend()
plt.show()
```

```python
df = pd.read_csv("Datasets/GOLD.csv", date_parser=['Date'])
df.Date = pd.to_datetime(df.Date)
df.head()
```

```python
result = df.copy()

y_req = result['new'][411:512].values.reshape(-1, 1)
r = lr.predict(y_req).ravel()
result['Pred'][411:512] = list(r)
```

```python
result.info()
```

```python
plt.figure(figsize=(20, 7))
sns.distplot(result['Pred'], color='blue', kde=True)
plt.title('Dist Plot', fontsize=18)
plt.xlabel('Pred Values', fontsize=15)
plt.show()
```

```python
### Importing Gail.csv as i chose GAIL stocks.
gail_df = pd.read_csv('Datasets/GAIL.csv')
gail_df['Date'] = pd.to_datetime(gail_df['Date'])
gail_df = gail_df.sort_values('Date')
gail_df.set_index('Date', inplace=True)
gail_df.head(3)
```

```python
gail_df.duplicated().sum()
```

```python
nifty_df = pd.read_csv('Datasets/NIFTY50.csv')
nifty_df['Date'] = pd.to_datetime(nifty_df['Date'])
nifty_df = nifty_df.sort_values('Date')
nifty_df.set_index('Date', inplace=True)
nifty_df.head(10)
```

```python
nifty_df.duplicated().sum()
```

```python
three_months_gail = gail_df[-90:]
three_months_nifty = nifty_df[-90:]
```

```python
three_months_gail['pct_change'] = three_months_gail['Close Price'].pct_change().copy()
three_months_nifty['pct_change'] = three_months_nifty['Close'].pct_change().copy()
```

```python
plt.figure(figsize=(20, 10))
three_months_nifty['pct_change'].plot()
three_months_gail['pct_change'].plot()
plt.title('Daily Returns of GAIL vs NIFTY')
plt.ylabel("Close Price")
plt.xlabel('Date')
plt.show()
```

```python
x = three_months_nifty['pct_change'].dropna()
y = three_months_gail['pct_change'].dropna()
```

```python
from statsmodels import api as sm

model = sm.OLS(x, y).fit() ### OLS Regression
print(model.summary())
```

```python
df = pd.concat([gail_df['Close Price'], nifty_df['Close']], axis=1)
df.columns = ['GAIL', 'NIFTY']
df.head()
```

```python
monthly_returns = df.pct_change(1).dropna(axis=0)
#clean_monthly_returns = monthly_returns  # drop first missing row
monthly_returns.head(10)
```

```python
x = monthly_returns['GAIL']
y = monthly_returns['NIFTY']
```

```python
X1 = sm.add_constant(x)
model = sm.OLS(y, X1)
```

```python
result = model.fit()
print(result.summary())
```

__The monthly Beta value for the stock GAIL is 0.0940__

which is less than 1 and hence it is less volatile than the benchmark
