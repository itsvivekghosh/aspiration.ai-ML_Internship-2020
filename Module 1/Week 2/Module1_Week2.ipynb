### Problem 1.5:
#### 1. Write a function to calculate the `average price` over the last N days of the stock price data where `N` is a user defined parameter.

#### 2. Write a second function to calculate the `profit/loss percentage` over the last N days.
Calculate the `average price` AND the `profit/loss percentages` over the course of last: 
1. 1 week,
2. 2 weeks,
3. 1 month,
4. 3 months,
5. 6 months and
6. 1 year.<br>

*Note : Profit/Loss percentage between `N days` is the percentage change between the closing prices of the 2 days.*

```python
def calculate_stock_price_n_days(days):
    
    data=df.tail(days)
    
    avg_sock_price = data['Average Price'].mean()
    
    return avg_sock_price

def calculate_profit_loss_percentage(days):
    
    data = df.tail(days)
    
    # price_value is the value of the difference between the closing prices of last two days: 
    price_value = data['Close Price'].iloc[days-1] - data['Close Price'].iloc[0]
        
    profit_loss_percentage = (price_value / data['Close Price'].iloc[days-1]) * 100
    
    return abs(profit_loss_percentage) ## calculating the absolute value
```

__We are taking 5 days as working days in a week as we do have data for `5 days each week` or `20 days a month`. So, by this report we can conclude that we have:__
1. 5 days a week as working days.
2. 10 days as 2 weeks as working days.
3. 20 days as 1 month as working days.
4. 20 * 6 days as 6 months as working days.
5. 20 * 12 days as working days in a Year.


```python
print("Stock Prices for last N days are : ")
print("Stock Price of last Week: {}".format(round(calculate_stock_price_n_days(5), 2)))
print("Stock Price of last 2 Weeks: {}".format(round(calculate_stock_price_n_days(10), 2)))
print("Stock Price of last Month: {}".format(round(calculate_stock_price_n_days(20), 2)))
print("Stock Price of last 3 Months: {}".format(round(calculate_stock_price_n_days(20*3), 2)))
print("Stock Price of last 6 Months: {}".format(round(calculate_stock_price_n_days(20*6), 2)))
print("Stock Price of last 9 Months: {}".format(round(calculate_stock_price_n_days(9*20), 2)))
print("Stock Price of last 1 Year: {}".format(round(calculate_stock_price_n_days(12*20), 2)))
print("Stock Price of last 2 Years: {}".format(round(calculate_stock_price_n_days(20*24), 2)))
```

```python
print("Loss/Profit Percentage for last N days are: ")
print("Last 1 week Loss/Profit percentage is: {}%".format(round(calculate_profit_loss_percentage(5),2)))
print("Last 2 weeks Loss/Profit percentage is: {}%".format(round(calculate_profit_loss_percentage(10),2)))
print("Last 1 month Loss/Profit percentage is: {}%".format(round(calculate_profit_loss_percentage(20),2)))
print("Last 3 months Loss/Profit percentage is: {}%".format(round(calculate_profit_loss_percentage(20*3),2)))
print("Last 6 months Loss/Profit percentage is: {}%".format(round(calculate_profit_loss_percentage(20*6),2)))
print("Last 6 months Loss/Profit percentage is: {}%".format(round(calculate_profit_loss_percentage(20*9),2)))
print("Last 1 year Loss/Profit percentage is: {}%".format(round(calculate_profit_loss_percentage(20*12),2)))
```

### Problem 1.7: Add another column 'Trend'

whose values are:
- `'Slight or No change'` for 'Day_Perc_Change' in between -0.5 and 0.5
- `'Slight positive'` for 'Day_Perc_Change' in between 0.5 and 1
- `'Slight negative'` for 'Day_Perc_Change' in between -0.5 and -1
- `'Positive'` for 'Day_Perc_Change' in between 1 and 3
- `'Negative'` for 'Day_Perc_Change' in between -1 and -3
- `'Among top gainers'` for 'Day_Perc_Change' in between 3 and 7
- `'Among top losers'` for 'Day_Perc_Change' in between -3 and -7
- `'Bull run'` for 'Day_Perc_Change' >7
- `'Bear drop'` for 'Day_Perc_Change' <-7

```python
for i in range(len(df)):
    
    if (df.iloc[i, -2] >=-0.5) & (df.iloc[i, -2] <=0.5):
        df["Trend"] = 'Slight or No change'
        
    if (df.iloc[i, -2] >=0.5) & (df.iloc[i, -2] <= 1):
        df["Trend"] = 'Slight positive'
        
    if (df.iloc[i, -2] >=-0.5) & (df.iloc[i, -2] <=-1):
        df["Trend"] = 'Slight negative'
        
    if (df.iloc[i, -2] >=1) & (df.iloc[i, -2] <=3):
        df["Trend"] = 'Positive'
    
    if (df.iloc[i, -2] >=-1) & (df.iloc[i, -2] <=-3):
        df["Trend"] = 'Negative'
        
    if (df.iloc[i, -2] >=3) & (df.iloc[i, -2] <=7):
        df["Trend"] = 'Among top gainers'
    
    if (df.iloc[i, -2] >=-3) & (df.iloc[i, -2] <=-7):
        df["Trend"] = 'Among top losers'
        
    if (df.iloc[i, -2] > 7):
        df["Trend"] = 'Bull run'
        
    if (df.iloc[i, -2] < -7):
        df["Trend"] = 'Bear drop'
df.head(3)
```

### Problem 1.8 Finding the `average` and `median` values

Find the `average` and `median values` of the column `'Total Traded Quantity'` for each of the types of `'Trend'`.

Hint : use `'groupby()'` on the `'Trend'` column and then calculate the average and median values of the column `'Total Traded Quantity'`.

```python
avg_trend_quality = df.groupby(df.Trend)['Total Traded Quantity'].mean()
median_trend_quality = df.groupby(df.Trend)['Total Traded Quantity'].median()

print('Average value of the column Total Traded Quantity for each of the types of Trend: \n', avg_trend_quality)
print('*' * 90)
print('Median value of the column Total Traded Quantity for each of the types of Trend: \n', median_trend_quality)
```
